# ðŸ”¹ ì‹œë¦¬ì¦ˆ ìƒì„± ì›ë¦¬
- ì‹œë¦¬ì¦ˆëŠ” **ê°’(value)** ê³¼ **ì¸ë±ìŠ¤(index)** ë¡œ êµ¬ì„±ëœ 1ì°¨ì› êµ¬ì¡°ìž„
- **ì¸ë±ìŠ¤ ìžë™ ìƒì„±**: ì§€ì •í•˜ì§€ ì•Šìœ¼ë©´ 0ë¶€í„° ì‹œìž‘í•˜ëŠ” ì •ìˆ˜ ì¸ë±ìŠ¤ê°€ ë¶€ì—¬ë¨
- **ì‚¬ìš©ìž ì§€ì • ì¸ë±ìŠ¤**: ì˜ë¯¸ ìžˆëŠ” ë ˆì´ë¸”ì„ ì§ì ‘ ë¶€ì—¬ ê°€ëŠ¥ (ì˜ˆ: ì›” ì´ë¦„)
- ì¤‘ìš”í•œ ì›ë¦¬: **ê°’ì€ ìœ„ì¹˜(ì ˆëŒ€ ì¸ë±ìŠ¤)ë¡œ ì ‘ê·¼ ê°€ëŠ¥í•˜ë©°, ë™ì‹œì— ë ˆì´ë¸”(ì˜ë¯¸ ì¸ë±ìŠ¤)ë¡œë„ ì ‘ê·¼ ê°€ëŠ¥**

```python
import pandas as pd

# 12ê°œì›” í‰ê·  ê¸°ì˜¨ ë°ì´í„° ìž…ë ¥
temp = pd.Series([-0.8, -0.1, 7.7, 13.8, 18.0, 22.4,
                  25.9, 25.3, 21.0, 14.0, 9.6, -1.4])

# ìžë™ ìƒì„±ëœ ì¸ë±ìŠ¤ í™•ì¸ (0~11)
print("ìžë™ ìƒì„± ì¸ë±ìŠ¤:\n", temp)

# ì›” ì´ë¦„ì„ ë ˆì´ë¸” ì¸ë±ìŠ¤ë¡œ ì§€ì •
temp.index = ['1ì›”','2ì›”','3ì›”','4ì›”','5ì›”','6ì›”',
              '7ì›”','8ì›”','9ì›”','10ì›”','11ì›”','12ì›”']

print("\nì‚¬ìš©ìž ì§€ì • ì¸ë±ìŠ¤:\n", temp)
```

# ðŸ”¹ ì‹œë¦¬ì¦ˆ ì •ë³´ í™•ì¸ ì›ë¦¬
- **size**: ì €ìž¥ëœ ê°’ì˜ ì´ ê°œìˆ˜  
- **shape**: ë°°ì—´ì˜ êµ¬ì¡° (íŠœí”Œ í˜•íƒœ, ì‹œë¦¬ì¦ˆëŠ” í•­ìƒ `(n,)`)  
- **dtype**: ê°’ì˜ ë°ì´í„° íƒ€ìž… â†’ ì—°ì‚° ì‹œ ì²˜ë¦¬ ë°©ì‹ì— ì˜í–¥  
- ì›ë¦¬: ì‹œë¦¬ì¦ˆëŠ” **ë„˜íŒŒì´ ë°°ì—´(ndarray)ì˜ í™•ìž¥íŒ**ì´ë¯€ë¡œ, ì´ëŸ¬í•œ ì†ì„±ë“¤ì„ ê·¸ëŒ€ë¡œ ê°€ì§

```python
print("ê°’ì˜ ê°œìˆ˜(size):", temp.size)
print("ë°°ì—´ì˜ í˜•íƒœ(shape):", temp.shape)
print("ë°ì´í„° íƒ€ìž…(dtype):", temp.dtype)
```

# ðŸ”¹ ì¸ë±ì‹±Â·ìŠ¬ë¼ì´ì‹± ì›ë¦¬
- **iloc**: ì ˆëŒ€ ìœ„ì¹˜ ê¸°ë°˜ (ì •ìˆ˜ ì¸ë±ìŠ¤) ì ‘ê·¼  
- **loc**: ì˜ë¯¸ ìžˆëŠ” ë ˆì´ë¸” ì¸ë±ìŠ¤ ì ‘ê·¼  
- ìŠ¬ë¼ì´ì‹±: ë²”ìœ„ ì§€ì • ì‹œ **ì‹œìž‘ í¬í•¨, ë ì œì™¸** (iloc) / **ì‹œìž‘~ë ëª¨ë‘ í¬í•¨** (loc)  
- ì›ë¦¬: í•˜ë‚˜ì˜ ë°ì´í„°ì— ëŒ€í•´ **ë‘ ê°€ì§€ ì¢Œí‘œ ì²´ê³„(ìœ„ì¹˜ vs ë ˆì´ë¸”)** ë¥¼ ë™ì‹œì— ì œê³µ

```python
# ì ˆëŒ€ ìœ„ì¹˜ ê¸°ë°˜ ì ‘ê·¼
print("iloc[2] (3ë²ˆì§¸ ê°’):", temp.iloc[2])

# ë ˆì´ë¸” ê¸°ë°˜ ì ‘ê·¼
print("loc['3ì›”']:", temp.loc['3ì›”'])

# ì—¬ëŸ¬ ê°œ ì„ íƒ
print("iloc[[3,5,7]]:", temp.iloc[[3,5,7]])
print("loc[['4ì›”','6ì›”','8ì›”']]:", temp.loc[['4ì›”','6ì›”','8ì›”']])

# ë²”ìœ„ ì§€ì •
print("iloc[5:8]:", temp.iloc[5:8])       # 6~8ì›”
print("loc['6ì›”':'9ì›”']:", temp.loc['6ì›”':'9ì›”'])  # 6ì›”~9ì›”
```

# ðŸ”¹ ì¡°ê±´ ê²€ìƒ‰ ì›ë¦¬
- ì‹œë¦¬ì¦ˆëŠ” ë„˜íŒŒì´ ë°°ì—´ì²˜ëŸ¼ **ì¡°ê±´ì‹ì„ ê·¸ëŒ€ë¡œ ì ìš©** ê°€ëŠ¥
- ì¡°ê±´ì— ë§žëŠ” ê°’ë§Œ True/Falseë¡œ êµ¬ë¶„ â†’ Trueì¸ ê°’ë§Œ ë°˜í™˜
- **where()**: ì¡°ê±´ ë§Œì¡±í•˜ì§€ ì•ŠëŠ” ê°’ì€ NaNìœ¼ë¡œ ì²˜ë¦¬ â†’ í•„ìš” ì‹œ dropna()ë¡œ ì œê±°
- ì›ë¦¬: **ë¶ˆë¦¬ì–¸ ì¸ë±ì‹±(Boolean Indexing)** í™œìš©

```python
# ì¡°ê±´ì‹ ì§ì ‘ ì‚¬ìš©
print("15ë„ ì´ìƒì¸ ë‹¬:\n", temp[temp >= 15])

# ì¡°ê±´ì‹ ì¡°í•© (and = &, or = |)
print("\n15~25ë„ ì‚¬ì´:\n", temp[(temp >= 15) & (temp < 25)])

# where() í™œìš©
print("\nwhere() ì‚¬ìš©:\n", temp.where(temp >= 15))
print("\nwhere() + dropna():\n", temp.where(temp >= 15).dropna())
```

# ðŸ”¹ ì—°ì‚°Â·í†µê³„ ì›ë¦¬
- ì‹œë¦¬ì¦ˆëŠ” **ë²¡í„°í™” ì—°ì‚°** ì§€ì› (ê° ì›ì†Œì— ì—°ì‚°ì´ ìžë™ ì ìš©ë¨)
- ë¸Œë¡œë“œìºìŠ¤íŒ…: ë™ì¼ ì—°ì‚°ì´ ì „ì²´ì— í™•ìž¥ ì ìš©
- í†µê³„ ë©”ì„œë“œ: í•©ê³„, í‰ê· , ìµœëŒ“ê°’, í‘œì¤€íŽ¸ì°¨ ë“± â†’ ë°ì´í„° ìš”ì•½ ê°€ëŠ¥

```python
# ì—°ì‚° ì˜ˆì‹œ
print("ëª¨ë“  ê°’ +1:\n", temp + 1)
print("2 * temp + 0.1:\n", 2 * temp + 0.1)

# í†µê³„ ë©”ì„œë“œ
print("í•©ê³„:", temp.sum())
print("í‰ê· :", temp.mean())
print("í‘œì¤€íŽ¸ì°¨:", temp.std())
print("ê¸°ì´ˆ í†µê³„ ì •ë³´:\n", temp.describe())
```

# ðŸ”¹ ìˆ˜ì •Â·ì¶”ê°€Â·ì‚­ì œ ì›ë¦¬
- **ìˆ˜ì •**: ê¸°ì¡´ ì¸ë±ìŠ¤ ì§€ì • â†’ ê°’ ê°±ì‹ 
- **ì¶”ê°€**: ì¡´ìž¬í•˜ì§€ ì•ŠëŠ” ì¸ë±ìŠ¤ ì§€ì • â†’ ìƒˆë¡œìš´ ê°’ ì‚½ìž…
- **ì‚­ì œ**: drop() ë©”ì„œë“œë¡œ íŠ¹ì • ì¸ë±ìŠ¤ ì œê±°
- **ë³µì‚¬**: copy()ë¡œ ë…ë¦½ì  ê°ì²´ ìƒì„± (ì›ë³¸ ì˜í–¥ X)

```python
# ì‹œë¦¬ì¦ˆ ìƒì„±
score = pd.Series([75, 80, 90, 60],
                  index=['KOR', 'ENG', 'MATH', 'SOC'])

# ê°’ ìˆ˜ì •
score.loc['ENG'] = 70
score.loc['MATH'] = 85

# ê°’ ì¶”ê°€
score.loc['PHY'] = 50

# ê°’ ì‚­ì œ
score = score.drop('SOC')

print("ìµœì¢… ì„±ì  ì‹œë¦¬ì¦ˆ:\n", score)

# ë³µì‚¬
score_copy = score.copy()
score_copy.loc['KOR'] = 100
print("\në³µì‚¬ë³¸ ìˆ˜ì • (ì›ë³¸ ìœ ì§€):\n", score_copy)
print("\nì›ë³¸:\n", score)
```
